```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# vader score data
data1 = {'patriots': 10.64, 'bills': 09.73, 'cowboys': 10.11, 'chiefs': 12.39, 'giants': 11.07, 'raiders': 4.51, 'chargers': 11.25, 'cardinals': 6.68, 'lions': 1.97, 'browns': 05.49}
names = data1.keys()
colors = ['#0021C3', '#001AD6', '#D7E5E8', '#FF0000', '#0059FF', '#000000', '#FFD904', '#FF3333', '#04A1E5', '#FE6A15']
temp = colors
# sort the dictionary by its values in ascending order
sorted_data1 = dict(sorted(data1.items(), key=lambda x: x[1]))

# sort the colors list to correspond to the sorted dictionary keys
sorted_colors = [colors[list(data1.values()).index(sorted_data1[key])] for key in sorted_data1.keys()]

labels = sorted_data1.keys()
values = sorted_data1.values()
colors = sorted_colors

vals = [int(i) for i in values]
print(np.mean(vals))


# Create bar chart
plt.bar(labels, values, color=colors)
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

# Add chart title and axis labels
plt.title('NFL Teams Sentiment Scores (Vader)')
plt.xlabel('Teams')
plt.ylabel('Vader Scores')

# Show chart
plt.show()
```

```python
# text blob score data
data2 = {'patriots': 11.12, 'bills': 8.85, 'cowboys': 8.66, 'chiefs': 11.00, 'giants': 9.75, 'raiders': 6.61, 'chargers': 10.18, 'cardinals': 5.09, 'lions': 7.32, 'browns': 6.21}
colors = temp
# sort the dictionary by its values in ascending order
sorted_data2 = dict(sorted(data2.items(), key=lambda x: x[1]))

# sort the colors list to correspond to the sorted dictionary keys
sorted_colors = [colors[list(data2.values()).index(sorted_data2[key])] for key in sorted_data2.keys()]
print(sorted_data2)
print(sorted_colors)
labels = sorted_data2.keys()
values = sorted_data2.values()
colors = sorted_colors

vals = [int(i) for i in values]
print(np.mean(vals))

# Create bar chart
plt.bar(labels, values, color=colors)
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

# Add chart title and axis labels
plt.title('NFL Teams Sentiment Scores (Text Blob)')
plt.xlabel('Teams')
plt.ylabel('Text Blob Scores')

# Show chart
plt.show()
```

```python
# Extract keys and values
values = [data1['patriots'] + data2['patriots'], data1['bills'] + data2['bills'], data1['cowboys'] + data2['cowboys'], data1['chiefs'] + data2['chiefs'], data1['giants'] + data2['giants'], data1['raiders'] + data2['raiders'],
            data1['chargers'] + data2['chargers'], data1['cardinals'] + data2['cardinals'], data1['lions'] + data2['lions'], data1['browns'] + data2['browns']]
print(values)
print(names)

colors = temp
# sort the values list by its values in ascending order
sorted_values, sorted_colors = zip(*sorted(zip(values, colors)))
values = sorted_values
print(sorted_values)
labels = ['lions', 'raiders', 'browns', 'cardinals', 'bills', 'cowboys', 'giants', 'chargers', 'patriots', 'chiefs']

colors = sorted_colors
vals = [int(i) for i in values]
print(np.mean(vals))

# Create bar chart
plt.bar(labels, values, color=colors)
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

# Add chart title and axis labels
plt.title('NFL Teams Sentiment Scores (Combined Scores)')
plt.xlabel('Teams')
plt.ylabel('Combined Scores')

# Show chart
plt.show()
```

```python
# Define x and y values for the points
wins = [4, 6, 7, 8, 9, 9, 10, 12, 13, 14, 15]
total_scores = [data1['cardinals'] + data2['cardinals'], data1['raiders'] + data2['raiders'], data1['browns'] + data2['browns'], data1['patriots'] + data2['patriots'], data1['lions'] + data2['lions'],
                data1['giants'] + data2['giants'], data1['chargers'] + data2['chargers'], data1['cowboys'] + data2['cowboys'], data1['bills'] + data2['bills'], data1['chiefs'] + data2['chiefs'], 25]
labels = ['Cardinals', 'Raiders', 'Browns', 'Patriots', 'Lions', 'Giants', 'Chargers', 'Cowboys', 'Bills', 'Chiefs', '']
colors = ['#FF3333', '#000000', '#FE6A15', '#0021C3', '#04A1E5', '#0059FF', '#FFD904', '#D7E5E8', '#001AD6', '#FF0000', '#FFFFFF']

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the points and label them
for x, y, label, color in zip(wins, total_scores, labels, colors):
    ax.plot(x, y, 'o', color=color)
    ax.annotate(label, (x + 0.1, y))

# Add a trend line
z = np.polyfit(wins, total_scores, 1)
p = np.poly1d(z)
ax.plot(wins,p(wins),"g--")

# Add labels and a title
ax.set_xlabel('Wins')
ax.set_ylabel('Combined Scores')
ax.set_title('Wins vs Combined Scores')

# Display the graph
plt.show()
```
